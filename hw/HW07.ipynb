# 1. Conceptual Definitions

### 1.1 The difference between Simple Linear Regression and Multiple Linear Regression; and the benefit the latter provides over the former:

Multiple Linear Regression (MLR) extends the concept of Simple Linear Regression by incorporating multiple predictor variables to model the outcome variable. The model is represented by an equation similar to the SLR model, but with additional terms for each variable:

$$ \large Y_i = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_i x_i$$

Here, we interpret $\beta_0$ as the intercept of the line modelling the **continuous** outcome variable. (Note that in alignment with the course notes, we will assume the $\epsilon$ term to be present without writing it, for simplicity).

MLR can offer a more realistic model by considering multiple factors that might influence an outcome; it can be unreasonable to assume just ONE predictor influences the outcome, so MLR takes the idea of SLR and extends it to multiple influences, and allows us to compare the strengths of those influences on the outcome.

We will discuss further categorical and continuous variations of the above equation in the following answers.


$$ \large Y_i = \beta_0 + \beta_1 1_\text{[group]}k_i + \beta_2 1_\text{[group]}k_i + \cdots + \beta_i 1_\text{[group]}k_i $$ 

__

### 1.2: the difference between using a continuous variable and an indicator variable in Simple Linear Regression; and these two linear forms:

The continuous form of Simple Linear Regression takes on the equation:

$$ \large Y_i = \beta_0 + \beta_1 x_1$$

The categorical indicator variable form of SLR takes on the equation:

$$ \large Y_i = \beta_0 + \beta_1 \times 1_\text{[group]} k_i $$

In the continuous variable model, the slope coefficient $\beta_1$ is the average change in the outcome variable for a one-unit increase in the **continuous** predictor variable. The $\beta_0$ term is the y-intercept of the line that is modelled; this is simply a vertical translation of the line, or can also be thought of as the value of $Y_i$ when $x_i = 0$.

In the indicator model, we deal with categorical predictor variables instead of continuous ones. Thus, we are no longer dealing in terms of slopes and intercepts, but baselines and contrasts. The baseline of the model is $\beta_0$, whereas the contrast is $\beta_1$. For a categorical variable, $k$, if $k_i$ satisfies the desired group (ie. $k_i \in \text{group}$), then the contrast variable $\beta_1$ is "turned on." So, the average of such variables is $\beta_0 + \beta_1$. Conversely, the average of predictor variables that don't satisfy the group condition have an average of $\beta_0$. So, the difference between these two averages is $\beta_1$; hence, the name "contrast" variable. 

So, the indicator model helps find the difference in categories of a predictor variable.

__

### 1.3: the change that happens in the behavior of the model (i.e., the expected nature of the data it models) when a single indicator variable is introduced alongside a continuous variable to create a Multiple Linear Regression; and these two linear forms (i.e., the Simple Linear Regression versus the Multiple Linear Regression):

When modelling an indicator variable with a continuous variable, we obtain the following equation:

$$ \large Y_i = \beta_0 + \beta_1 \times 1_\text{[group]} k_i + \beta_2 z_i $$

In this model, we have a combined explanation of categorical comparison and continuous effects. That is, we still retain the same properties in which $\beta_0$ is the ***baseline***, but now we have two different types of coefficients. In the aboce equation, we have a categorical variable $k_i$, in which a $k_i$ satisfying a given group is applied a contrast of $\beta_1$. However, we now also have a slope coefficient of $\beta_2$. 

This means that we form two parallel lines with the same slope $\beta_2$. The lines will differ by a vertical translation of $\beta_1$, since this is the contrast between the two groups being studied. We can extend this idea to more variables, as outlined below.

### 1.4: the effect of adding an interaction between a continuous and an indicator variable in Multiple Linear Regression models; and this linear form:

The model in 1.3 accounts for a constant slope between groups, producing parallel lines. But, what if we wanted to see the continuous effect of a group on the outcome? We may add the below interaction:

$$ \large Y_i = \beta_0 + \beta_1 x_i + \beta_2 \times 1_{[\text{group}]}(k_i) + \beta_3 \times x_i \times 1_{[\text{group}]}(k_i)$$

Thus, by adding the indicator "switch" to a multiplication of a slope coefficient, we can adjust each group's slope accordingly. 

For variables not satisfying the group condition (ie. $k_i = 0$), we get $Y_i = \beta_0 + \beta_1 x_i$, 
Otherwise, for variables satisfying the group condition, we get $ Y_i = \beta_0 + \beta_1 x_i + \beta_2 \times 1_{[\text{group}]}(k_i) + \beta_3 \times x_i \times 1_{[\text{group}]}(k_i)$.

The first line has intercept $\beta_0$ and slope $\beta_1$. The second line has intercept $\beta_0 + \beta_2$ and slope $\beta_1 + \beta_3$.

Thus, this interaction allows us to have differnet slopes depending on the group term, providing a dynamic modelling of predictors to outcomes.

### 1.5: the behavior of a Multiple Linear Regression model (i.e., the expected nature of the data it models) based only on indicator variables derived from a non-binary categorical variable; this linear form; and the necessarily resulting binary variable encodings it utilizes:

The previous answers addressed binary categorical variables, in which we either satisfy a condition, or we don't. For categories with multiple responses, we can actually employ the same methodology. So, for one "group" of categories, we have one choice of the group as the baseline, with the rest of the group as the references to that baseline (meaning we have to add indicator variables to make this reference). 

The following is the linear form of the model:

$$Y_i = \beta_0 + \sum_{j=1}^{K-1} \beta_j \times 1_{[\text{Category j}]}(k_i)$$

As we see below, we have K-1 indicator variables, where K is the number of categories in a "group" of predictor variables. The "unaccounted" variable is represented by $\beta_0$. Thus, every other category is a reference to this baseline. This is done by this "binary encoding" that flips an associated $k_i$ **on or off** (binary) depending on the satisfaction of a condition.

# 2. Multiple Linear Regression Example Discussion

### 2.1 

Let Sales be the outcome variable, predicted by continuous variables TV and Online, which we will call $x$ and $z$ respectively.

Without any interaction between variables, we obtain the following model from the scenario:

$$ \text{Sales} = \beta_0 + \beta_1 x_i + \beta_2 z_i $$

Thus, we observe that sales are predicted independently and ***continuously*** by each predictor variable. That is, the coefficients of each predictor is unaffected by the other. 

However, if we wanted an interaction between the two, we could have a "weighted" affect by adding an additional term that quantifies the combined effect of both predictor variables at the same time:

$$ \text{Sales} = \beta_0 + \beta_1 x_i + \beta_2 z_i + \beta_3 x_i z_i$$

Intuitively, we can characterize this 3rd term as being a syngergistic effect between the two variables; if $x_i$ is large, but $z_i$ is small, then the effect of $x_i$ is diminished (and vice versa). But if both variables are large, their combined effect will be even greater.

### 2.2

In the case of categorical variables, we may categorize the predictor variables as being "High" or "Not High" (ie. low) for the purpose of creating binary varibales. The linear form of this model becomes:

$$ \large \text{Sales} = \beta_0 + \beta_1 \times 1_\text{[High]} x_i + \beta_2 \times 1_\text{[High]}z_i$$

In this case, $\beta_0$ is the baseline, which occurs when TV and Online budgets are BOTH low. 

When TV budget is high, but Online budgets are low, we add the contrast $\beta_1$, for an average of $\beta_0 + \beta_1$. 

When Online Budget is high, but TV budget is low, we add the contrast $\beta_2$, for an average of $\beta_0 + \beta_2$. 


However, this isnt particularly nuanced, as each variable is independent and does not affect the other. We can add a synergestic interaction between them by the following equation:

$$ \large \text{Sales} = \beta_0 + \beta_1 \times 1_\text{[High]} x_i + \beta_2 \times 1_\text{[High]}z_i + \beta_3 \times 1_\text{[High]} x_i z_i$$.

Similarly to the explanation in 2.1, the effect here is synergestic in the last term. This means that the effects of the variables are combined. Since we're dealing with categorical variables, for the last term to be activated BOTH TV and Online budgets must be high, but this model allows us to see WHAT effect this has in this case (which is intuitively more telling...).

# 3. CSCH Logistic Regression

### 3.1.

Firstly, we will create an additive linear (logistic) regression based on the following variables.

**Outcome:** WELLNESS_self_rated_mental_health_binary (a binary variable which we encode to either be "good" or "bad").

**Predictors**: DEMO_age (continuous age variable), CONNECTION_community_barriers_introvert_binary (binary indicator variable of self-reported introvert vs. extrovert), CONNECTION_community_involvement_sports_p3m_binary (binary indicator variable reporting a person's sports activity level).

If this were linear regression, it would look like this:

$$ \large Y_i = \beta_0 + \beta_1 z_\text{age} + \beta_2 \times 1_\text{["introvert"]}k_i + \beta_3 \times 1_\text{["sporty"]}k_i$$

The below code sets up and creates our model.

```python
import pandas as pd
cols = pd.read_csv("../course_project/data/var_names.csv")

data = pd.read_csv("../course_project/data/CSCS_data_anon.csv", 
                   na_values=["9999", "", " ", "Presented but no response", "NA"],
                   low_memory=False)  # Specify the desired type


empty = (data.isna().sum()==data.shape[0])
data = data[empty.index[~empty]] # keep non empty columns only

data = data[data.REMOVE_case=='No']

data = data[data.SURVEY_collection_year==2022]

data.shape
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np

# Binary mapping for WELLNESS_self_rated_mental_health
data.loc[:, 'WELLNESS_self_rated_mental_health_binary'] = data['WELLNESS_self_rated_mental_health'].map({
    'Poor': 0,
    'Fair': 0,
    'Good': 1,
    'Very good': 1,
    'Excellent': 1
})

# Define binary mapping for the introvert indicator
introvert_mapping = {
    'I am an introvert': 1,
    'Not Selected': 0,
    np.nan: 0  # Convert NaN values to 0
}

# Apply mapping to create binary column
data.loc[:, 'CONNECTION_community_barriers_introvert_binary'] = \
    data['CONNECTION_community_barriers_introvert'].map(introvert_mapping)

# Define the mapping for sports involvement
sports_mapping = {
    "Yes": 1,
    "No, but I would like to": 0,
    "No, and I'm not interested in doing so": 0,
    np.nan: 0  # Treat NaNs as "not interested"
}

# Apply the mapping to create the binary column
data.loc[:, 'CONNECTION_community_involvement_sports_p3m_binary'] = data['CONNECTION_community_involvement_sports_p3m'].map(sports_mapping)

# Ensure there are no NaNs in the outcome variable and predictors
data = data.dropna(subset=['WELLNESS_self_rated_mental_health_binary', 'CONNECTION_community_involvement_sports_p3m_binary', 'CONNECTION_community_barriers_introvert_binary', 'DEMO_age'])
```

```python

# Define the logistic regression formula
formula_additive = 'WELLNESS_self_rated_mental_health_binary ~ DEMO_age + CONNECTION_community_barriers_introvert_binary + CONNECTION_community_involvement_sports_p3m_binary'

# Fit the logistic regression model
model_additive = smf.logit(formula=formula_additive, data=data).fit()

# Print the model summary
print(model_additive.summary())
```

### 3.2.
Leaving the above for later interpretation, we move on to a synergistic form of regression.

The form of this model will have an interaction between (1) age and reported introversion, as well as (2) age and reported sport activity.

The below equation models this:

$$ \large Y_i = \beta_0 + \beta_1 z_{\text{age}} + \beta_2 \times 1_{\text{["introvert"]}}k_i + \beta_3 \times 1_{\text{["sporty"]}}k_i + \beta_4 \left( z_{\text{age}} \times 1_{\text{["introvert"]}}k_i \right) + \beta_5 \left( z_{\text{age}} \times 1_{\text{["sporty"]}}k_i \right)
$$

```python
# Synergistic model formula with main effects and interaction terms
formula_synergistic = (
    'WELLNESS_self_rated_mental_health_binary ~ '
    'DEMO_age + CONNECTION_community_barriers_introvert_binary + '
    'CONNECTION_community_involvement_sports_p3m_binary + '
    'DEMO_age:CONNECTION_community_barriers_introvert_binary + '
    'DEMO_age:CONNECTION_community_involvement_sports_p3m_binary'
)

# Fit the model
model_synergistic = smf.logit(formula=formula_synergistic, data=data).fit()

# Display the model summary
print(model_synergistic.summary())
```

### 3.3.

As noted above, the additive model is as follows:
$ \;\; Y_i = \beta_0 + \beta_1 z_\text{age} + \beta_2 \times 1_\text{["introvert"]}k_i + \beta_3 \times 1_\text{["sporty"]}k_i$

**Intuitive Explanation:**
- $\beta_0$: The intercept, or baseline, outcome when age = 0, and a person is not introverted or sporty.
- $\beta_1$: The slope coefficient of age; measures the rate of change of the outcome for each 1-year increase in age.
-$\beta_2$: The contrast term between introverts and extroverts, levaraging binary encoding (purely additive; has no effect on age/sportiness).
- $\beta_3$: The contrast term between sporty vs non-sporty people.

As noted above, the synergistic model is as follows:
$ Y_i = \beta_0 + \beta_1 z_{\text{age}} + \beta_2 \times 1_{\text{["introvert"]}}k_i + \beta_3 \times 1_{\text{["sporty"]}}k_i + \beta_4 \left( z_{\text{age}} \times 1_{\text{["introvert"]}}k_i \right) + \beta_5 \left( z_{\text{age}} \times 1_{\text{["sporty"]}}k_i \right)
$

**Intuition:**
- This has the same interpretation as the additive model, but also assumes interaction between predictors. 
- As explained in Q2, these interactions result in one predictor having a dependent impact on the effect of another. 
- In the synergistic model, being both "sporty" and "introverted" might have a stronger or weaker combined effect, not just the sum of their individual effects.

## Conceptual Explanations

We often treat p-values and $R^2$ as similar evidences of a combined purpose in the context of regression analysis. However, it is not true that both of them tell the same story, or that they both must have a "statistically significant" result to "confirm" each other. 


The apparent contradiction between the two statements comes from the different things that **R-squared** and **p-values** represent in regression analysis.

**R-squared** shows how much of the variability in the outcome variable is explained by the model. In this case, a value of **17.6%** means that the model explains only a small portion of the total variability in the data. While the model does capture some key relationships, it suggests that other factors not included in the model may also play a significant role in affecting the outcome.

**P-values**, on the other hand, test whether the effect of each predictor variable is statistically significant. A **low p-value** indicates that there is strong evidence against the null hypothesis (which suggests no effect), meaning the predictor has a **meaningful relationship** with the outcome, controlling for other variables in the model.

However, ***these two concepts ARE NOT MUTUALLY EXCLUSIVE***.

These concepts don't conflict; instead, they focus on different aspects of the model:

- **R-squared** tells us about the overall goodness of fit—**how well the model explains the variation in the outcome**. 

- **P-values**, however, show us the strength of evidence for the relationship between each predictor and the outcome, **independent of the model’s overall explanatory power**.

It’s completely possible for a model to have a low **R-squared** (indicating limited explanatory power) and still have **statistically significant coefficients** with **large effect sizes**. This means that while the predictors in the model have **strong individual effects** on the outcome, there could be other factors outside of the model that account for a significant portion of the outcome's variability. So, **both measures can be interpreted meaningfully together** because they focus on different aspects of the model's performance and the predictors' relationships with the outcome.

## Example Explanation

- **R-squared = 17.6%** means the model explains only a small portion (17.6%) of the variability in the outcome, **suggesting there are other unmeasured factors** contributing to the remaining 82.4% of the variation.

- **Large coefficients with low p-values** indicate that the predictors included in the model have a **significant relationship** with the outcome and **contribute meaningfully** to predicting the outcome variable.

In summary, the model captures some of the variability in the outcome, but **not enough to fully explain it**. Despite the low **R-squared**, the predictors are still **important** and have a **strong effect** on the outcome, as shown by their **large coefficients** and **low p-values**.

This isn't contradictory because **R-squared** and **p-values** measure different aspects of the model: **R-squared** shows how much of the outcome's variability is explained by the model, while **p-values** indicate the significance of each predictor's relationship with the outcome, regardless of the model's overall explanatory power.

This step defines a linear regression model using the formula `'HP ~ Attack + Defense'`, where `HP` is predicted based on `Attack` and `Defense` from the training data (`pokeaman_train`). The `fit()` function is used to estimate the coefficients of the predictors. After fitting the model, a summary is generated, which provides the coefficients for `Attack` and `Defense`, R-squared (indicating how well the model fits the data), and p-values (to test the statistical significance of the predictors).


```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

The model's **R-squared** value is 0.148, meaning it explains only 14.8% of the variation in `HP`, which is quite low. This suggests other factors are affecting `HP` that aren’t captured by the model. The **Adjusted R-squared** is slightly lower at 0.143, indicating that adding more predictors doesn’t significantly improve the model's explanatory power.

The **coefficients** show that:
- The **Intercept** is 42.5882, meaning the baseline `HP` when both `Attack` and `Defense` are zero is 42.5882.
- For **Attack**, each additional unit increases `HP` by 0.2472. This predictor is statistically significant with a very low p-value (0.000).
- For **Defense**, each additional unit increases `HP` by 0.1001, and it’s also statistically significant with a p-value of 0.028.

Both **Attack** and **Defense** are statistically significant predictors of `HP`, but the low R-squared value shows the model doesn’t explain a lot of the variation in `HP`, indicating there might be other important factors missing from the model.


## Cell 3

This step calculates both the **"in sample"** and **"out of sample"** R-squared values to compare the model’s performance.

1. **Prediction on Test Data**: The model generates predictions for the test data (`pokeaman_test`), representing how well it predicts unseen `HP` values.

2. **"In sample" R-squared**: This value is calculated from the training data (`pokeaman_train`) and measures how well the model fits the training data. It was previously found to be 0.148.

3. **"Out of sample" R-squared**: This is calculated by comparing the predicted values to the actual `HP` values in the test data. It shows how well the model generalizes to new data.

4. **Print Statements**: The first prints the "in sample" R-squared, and the second prints the "out of sample" R-squared.

### Purpose:
This step helps assess the model’s performance by comparing its fit to the training data ("in sample") versus how well it generalizes to new data ("out of sample"). A large difference between the two could indicate overfitting.

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```


1. **In sample R-squared**: 0.1477 — The model explains about 14.77% of the variation in `HP` for the training data, which reflects a relatively poor fit.

2. **Out of sample R-squared**: 0.2121 — The model explains about 21.21% of the variation in `HP` for the test data, which is better than its fit to the training data.

### Takeaway:
- The **out of sample** R-squared is higher than the **in sample** R-squared, which is unusual. Typically, we'd expect worse performance on unseen data if the model is overfitting. However, in this case, it suggests the model might generalize better than it fits the training data, potentially due to issues with the representativeness of the training set or other influencing factors.

## Cell 4

This cell sets up and fits a **multiple linear regression model** that includes **interaction terms** between predictors:

1. **Formula**:
   - The model is specified with several predictors (`Attack`, `Defense`, `Speed`, etc.) and their **interaction effects** (e.g., how `Attack` and `Defense` jointly affect `HP`). 
   - The `Q()` function ensures that variables with special characters (like `Sp. Def` and `Sp. Atk`) are properly interpreted.
   - The commented-out part would add complex interactions for categorical variables, but it's avoided due to the large number of combinations that could make the model too complex.

2. **Model Specification**:
   - The model is created using `ols()` from **statsmodels**, which fits an ordinary least squares regression.

3. **Model Fitting**:
   - `model4_spec.fit()` fits the model to the training data.

4. **Model Summary**:
   - `model4_fit.summary()` provides detailed output, including coefficients and diagnostic statistics.

### Purpose:
- This model aims to capture more complex relationships by including both main effects and interactions between predictors, offering a more flexible fit for `HP`.


```python

model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

- **Overfitting Risk**: The model is likely overfitting the training data due to its complexity. It includes a large number of predictors (62 coefficients) and interaction terms that may capture noise rather than true relationships. A significant increase in R² could mean the model is capturing more nuances in the training data, but it could also mean it's overfitting to noise or specific patterns that won’t generalize well to new data.
  
- **Key Indicators**:
  - The model shows a large gap between **R-squared** (0.467) and **Adjusted R-squared** (0.369), suggesting the added complexity isn't improving the model's generalizability.
  - Many predictors, including interaction terms, are statistically significant, but these relationships may not be meaningful beyond the training data.
  - High **AIC** and **BIC** values (3603 and 3855) also signal that the model might be too complex.

- **Possible Solutions**:
  - **Cross-validation** can help assess the model's performance on unseen data.
  - Reducing the number of predictors, particularly interactions, could simplify the model and improve its ability to generalize.

### Key Insights:

## Explanation of Analysis

In `model4`, the linear model is specified with a high level of complexity, adding many predictor variables through interaction terms between centered and scaled features like `Attack`, `Defense`, `Speed`, `Sp. Def`, `Sp. Atk`, and the indicator variable `Legendary`. This design leads to an expanded "design matrix" (`model4_spec.exog`), where each interaction term occupies its own column, increasing the number of variables significantly. This increase in dimensionality introduces severe multicollinearity, which is confirmed by examining the correlation matrix of the design matrix (`np.corrcoef(model4_spec.exog)`) and the extremely high condition number for `model4_spec.exog`. Initially, the condition number is around 12 trillion, and even with centering and scaling, it still rises to 2.25 quadrillion—both indicators of severe multicollinearity.

The presence of such high multicollinearity makes the coefficient estimates unstable, which means they are highly sensitive to small changes in the data. As a result, `model4` is essentially overfitting, picking up on random noise in the training data instead of meaningful patterns that could generalize to new data. This overfitting is reflected in the model’s performance: although the in-sample R-squared (how well the model fits the training data) appears reasonable, the out-of-sample R-squared (how well the model performs on new, unseen data) drops dramatically, indicating that the model cannot generalize.

This sharp difference between the in-sample and out-of-sample R-squared values shows how an overly complex model, like `model4`, is prone to severe overfitting due to high multicollinearity. It ends up capturing coincidental relationships in the training data that do not apply outside of it. By comparison, a simpler model, such as `model3`, which avoids excessive interactions and instead focuses on more reliable, simpler relationships, is able to generalize better and shows more stable performance across different data.
