## 2. Distinctions Between Population Parameters and Sample Statistics in Hypothesis Testing

####  "It is important to note that outcomes of tests refer to the population parameter, rather than the sample statistic! As such, the result that we get is for the population."

When conducting a hypothesis test, we are attempting to arrive at a conclusion of the population. In other words, we are trying to describe something about a population parameter, $\mu$, based on the rejection of $\mu_0$, or lack thereof.

So, we construct a null hypothesis, $\mu_0$. If we reject this null hypothesis, this tells us the population parameter, $\mu$, is most probably described by $H_A$. 

To arrive at these conclusions, we make use of synthetic samples to calculate statistics that make up a theoretical distribution under the null hypothesis. HOWEVER, we are not making any conclusions on $\bar{x}$ or $x_i$, as these are sample statistics (which only concern a small portion of the population), when we are trying to make a conclusion about the larger population. We may use $x_i$ in the theoretical distribution if we are trying to make a conclusion about $\mu$, but nothing more.

Important definition recaps for each notation:

- $x_i$: An observation of a sample, where $i=1, 2, \cdots, n$ denotes the row/observation.
- $\bar{x}$: The sample mean  Given a SAMPLE of a population, this denotes the mean of the sample.
- $\mu$: This is the true population mean.
- $\mu_0$: This represents the population mean UNDER THE ASSUMPTION of the null hypothesis. That is, we are not concluding that this definitively represents the population parameter, but that this is what the null hypothesis suggests.

We use **statistics** from samples to estimate **parameters** of populations. Hypothesis tests use sample data to evaluate claims (**null hypotheses**) about population parameters. If a hypothesis test leads us to reject the null hypothesis, it suggests that the effect or difference observed in the sample likely applies to the entire population.&#x20;

## 3. "Imagine a world where the null hypothesis is true" when calculating a p-value... <br> Explain why this is in your own words in a way that makes the most sense to you.

Firstly, we should reinforce what a p-value is trying to describe. A p-value is the probability that a statistic is as or more extreme than the observed statistic under the assumption of the null hypothesis. In other words, we are trying to see the plausability of results obtained when we assume the null hypothesis. Recall that the entire purpose of hypothesis testing it to try and reject the null hypothesis, or fail at doing so. Thus, we "imagine" such a world where $H_0$ is true, and see if our observations in this world are ridiculous or not.

For example, if the null hypothesis claimed that 50% of soccer players were right footed, and our observed statistic stated that 85% of soccer players were right footed, we would end up with a "simulated $H_0$ world" that does not resemble what our observed statistic states. In this case, the "null world" would favour a normal distribution centred at 0.5, which leaves our observed statistic of 0.85 more extreme than much of the distribution. And so, seeing that this "null world" produces ridiculous results, we may reject it.

## 4.  Smaller p-values Makes the Null Hypothesis Less Plausable

This was somewhat explained in Q3, but here is a further explanation. Under the assumption of the null hypothesis, a lower p-value indicates that the null sampling distribution has a greater proportion of statistics less extreme than the observed test statistics, indicating that the observed test statistic is an unlikely result of this distribution (which we know to be ridiculous since we OBSERVED this statistic). 

More formally, we may analyze the process of obtaining a low p-value, and what this entails. To begin a hypothesis test, we assume the null hypothesis. Then, we create a large amount (perhaps 10,000) of SYNTHETIC samples under this value of $\mu_0$. From this, we create a sampling distribution, which should be centred at $\mu_0$. Then, we mark where our observed statistic is on either side of $\mu_0$ (assuming a two-tailed test). In this scenario, a low p-value would mean that a majority of the distribution's area is NOT as or more extreme than the observed statistic, meaning that the portion of the distribution past the observed statistic is an unlikely result under the null hypothesis. The reason this is ridiculous is because we know this observed statistic is a realistic outcome based on our original sample, and so if the null distribution is claiming otherwise, then we may disregard it.

For example, if I had a p-value of 0.01, this means that just 1% of the null distribution was at least as extreme as the observed statistic, which is ridiculous since the observed statistic is a realistic outcome. Below is a graph that illustrates this:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the null hypothesis sampling distribution
mu_0 = 0.5  # Null hypothesis mean (50% of students scoring at or below)
sigma = 0.1  # Standard deviation (adjust for clarity)
observed_stat = 0.75  # The observed statistic (75% of students scoring above)
symmetric_stat = 0.25  # Symmetric counterpart for two-tailed test (25% scoring below)
p_value = 0.01  # Example low p-value

# Create a range of x-values for the null hypothesis sampling distribution
x = np.linspace(mu_0 - 4 * sigma, mu_0 + 4 * sigma, 1000)
y = norm.pdf(x, mu_0, sigma)  # Probability density function for normal distribution

# Plot the null distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Null Hypothesis Distribution", color="blue")

# Shade the area past the observed statistic on the right (right tail)
x_fill_right = np.linspace(observed_stat, mu_0 + 4 * sigma, 1000)
y_fill_right = norm.pdf(x_fill_right, mu_0, sigma)
plt.fill_between(x_fill_right, y_fill_right, color="red", alpha=0.5, label='Right Tail')

# Shade the area past the symmetric statistic on the left (left tail)
x_fill_left = np.linspace(mu_0 - 4 * sigma, symmetric_stat, 1000)
y_fill_left = norm.pdf(x_fill_left, mu_0, sigma)
plt.fill_between(x_fill_left, y_fill_left, color="red", alpha=0.5, label='Left Tail')

# Shade the inner region (between the symmetric and observed statistics)
x_fill_inner = np.linspace(symmetric_stat, observed_stat, 1000)
y_fill_inner = norm.pdf(x_fill_inner, mu_0, sigma)
plt.fill_between(x_fill_inner, y_fill_inner, color="orange", alpha=0.5, label='Inner Region')

# Mark the observed statistic and its symmetric counterpart
plt.axvline(observed_stat, color="black", linestyle="--", label="Observed Statistic (75%)")
plt.axvline(symmetric_stat, color="black", linestyle="--", label="Symmetric Statistic (25%)")
plt.axvline(mu_0, color="black", linestyle="--", label="Null Hypothesis Mean (50%)")

# Add labels and title
plt.title("Test Scores Example: Two-Tailed Test of Null Hypothesis", fontsize=14)
plt.xlabel("Proportion of Students Scoring Above Threshold", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()

# Display the plot
plt.show()
```

If I observe that 75% of students in a sample passed, and want to reject the null hypothesis that 50% of all students passed, this would be the null distribution I would see. Here, the p-value would be the proportion of red-shaded area to the entire area of the graph, which would be a very small number. This makes the null hypothesis ridiculous because we know that 75% of students in a sample passed, while this null distribution indicates that this result is not plausible, when we have already observed it. This is a similar mindset to confidence intervals; the null distribution is claiming the red area is not favoured under this hypothesis, but here we KNOW this red area is plausible, so we reject this null hypothesis with a low p-value.

## 5. Gntrkn Kissing Couples Hypothesis

Below, I will implement a simulation of Gntrkn's hypothesis on kissing couples tilting their heads. As a brief explanation of the code below, we will start with our hypotheses:

- $H_0: \mu = 0.5$ (humans have no tendency to left/right tilt  ie. proportion is 50 Right - 50 Left)
- $H_0: \mu \neq 0.5$ (humans DO have a tendency to left/right tilt)

Thus, we run a synthetic sample based on the null hypothesis, $\mu_0$ (0.5), in order to make a conclusion about $\mu$. The means of these synthetic samples, $\bar{x}$, will be calculated by each iteration of a simulation by finding the average of all observations, $x_i$ (Note: there are 124 observations in each sample based on the original record). Then, we see how many of these $\bar{x}$ are as or more extreme than the observed test statistic of 0.645. The proportion of extreme means to the number of simulations will be the p-value:
- $p = \dfrac{\textit{# of extreme sample means}}{\textit{# of simulations}}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(1)

# Parameters for the simulation
mu_0 = 0.5  # Null hypothesis proportion (no tilt preference)
obs_stat = 0.645  # Observed proportion of right tilts (64.5%)
num_of_sim = 10000  # Number of simulations to run
num_of_ppl = 124  # Total number of couples in the study

# Array to store simulated proportions of right tilts
sim_proportions = np.zeros(num_of_sim)

# Simulate the proportions of right tilts for the null hypothesis
for i in range(num_of_sim):
    # Randomly choose 0 (left tilt) or 1 (right tilt) for each individual
    sim_choices = np.random.choice([0, 1], size=num_of_ppl, replace=True)
    # Calculate and store the mean (proportion of right tilts) for this simulation
    sim_proportions[i] = sim_choices.mean()

# Determine how many simulated proportions are as extreme or more extreme than the observed statistic
num_of_as_extreme = np.sum(np.abs(sim_proportions - mu_0) >= np.abs(obs_stat - mu_0))

# Calculate the p-value as the proportion of extreme results
p_value = num_of_as_extreme / num_of_sim

# Print the calculated p-value
print(f"P-value: {p_value:.4f}")
```

Based on the below table, we see that our p-value of 0.0015 indicates that we have strong evidence against the null hypothesis as our p-value lies in the interval of $(0.01, 0.001)$.


|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

Thus, due to our strong evidence against $H_0$, we may reject the null hypothesis. This implies that humans likely do have a tendency to tilt their heads, whether it be left or right leaning. By this hypothesis test, we do not conclude WHAT that tendency IS, but rather what it likely ISN'T.

## 6. Certainty of P-Values in "Proving" or "Disproving" Hypotheses

To sum up the following questions simply, a p-value will NEVER provide a definitive proof of anything.

### Q: Can a smaller p-value definitively prove that the null hypothesis is false?
NO! A p-value is the the probability that a statistic is as or more extreme than the observed statistic if the null hypothesis is true. We can use the value to characterize the strength of evidence against the null hypothesis. HOWEVER, no matter how improbable a null hypothesis may be, it always has a **possibility**. Thus, it must always have some **probability**, even if it is very negligent. A small p-value simply means we have a lot of evidence against the null hypothesis, but this does not translate to a PROOF.

<br>

### Q:  Is it possible to definitively prove that Fido (from the "second pre-lecture video") is innocent using a p-value?
Again, NO! There is always a possibility that Fido is guilty or innocent, which means there is an associated probability for those parameters as well. No matter the strength of evidence we have, we can never definitively say whether or not Fido is innocent. A proof would imply that we have a 100% accurate depiction of the population, which is not possible unless we know the population beforehand.

<br>

### Q: Is it possible to difinitively prove that Fido is guilty using a p-value?
Following the same discussion from the previous Q, NO! p-values are not proofs, they are tools used as evidence. The lower the p-value, the stronger our evidence against the null-hypothesis is, and the more we are confident that Fido is guilty. HOWEVER, this confidence is not to be mistaken with certainty. There is still room for Type-I and Type-II errors based on the fact that we could always get a false-positive/negative depending on the test conducted. So, we can never definitively prove if Fido is guilty with a p-value.

<br>

### Q: How low or high does a p-value have to be to definitely prove one or the other?
As a summary of the previous Q's, no matter how high or low a p-value is, we can never definitively prove Fido's innocence or guilt. A p-value is used as a measure of our evidence against $H_0$, but is never plausible to be used as a proof. There is always the possibility that Fido is innocent or guilty, so using a p-value as a proof is ridiculous.

## 8. Prof. Schwartz's Tea Experiment

### Problem Introduction
In the early 1920s, a stastician named Ronald Fisher and a colleague Dr. Muriel Bristol, were having a cup of tea. Bristol claimed that she disliked Fisher pouring the milk in before the tea, and that she could tell the difference. As a statistician would naturally do, Fisher conducted a test, making 8 cups of tea, with half being milk-first, and the other half being tea-first. Bristol correctly guessed all 8 cups. Being the skeptic he was, Fisher decided to conduct a hypothesis test and seek further confirmation on his suspicions.

Inspired by this famous statistical example of Fisher, Professor Schwartz wants to recreate this problem, by giving 80 of his STA130 students a cup of tea, afterward asking them if they could tell if the milk or tea was poured in first, without them having any prior knowledge to it. Prof. Schwartz has intentionally set up a sample that may be used to conduct a hypothesis test!

Of his 80 students, 49 had gotten their guesses correct. Note that the sample in this test is the 80 student sample, whereas the population would be extending this to all STA130 students (or people, for that matter) and their guesses.

Prof. Schwartz specifically wants to see that, on average (arithmetic mean $\mu$), do the students have a tendency to guess their cup right or wrong. The below statements are the hypotheses within this test:

- $H_0: \mu = 0.5$ (The students, on average, guess randomly and have no tendency to guess right or wrong).
- $H_A: \mu \neq 0.5$ (The students, on average, do have a tendency to guess right ***or*** wrong). 

The $\alpha$-significance level used for this test will be 0.05 (meaning $p<0.05$ to formally reject $H_0$).

### Quantitative Analysis

#### Methodology
- We will be performing a two-tailed test (as per $H_0$ and $H_A$), simulating a world of synthetic samples under the assumption of $H_0$. 
    - Number of Sim: 10,000
- Within these simulations, we will create samples of size 80 (original size), and calculate the sample mean ($\bar{x}$) of each sample, leaving us with 10,000 sample means.
- With these sample means, we will examine how many of them are ***as or more extreme than the observed test statistic***, in order to calculate the proportion of the p-value. "As or more extreme" in a two-tailed test implies that we are checking to see if a given simulated statistic is at least as far away as the observed statistic is from $\mu_0 = 0.5$ (symmetric on both ends).
    - What is our observed test statistic? That would be the sample mean of the original sample Prof. Schwartz conducted! That is, 49/80, or 0.6125.
- Thus, the p-value will be the proportion of these extreme sample means to the number of simulations (10,000):
    - $p = \dfrac{\textit{# of extreme test stats}}{10,000}$
    
    
<br>
<br>

We conclude that, if $p < \alpha = 0.05$, then we may reject $H_0$. Else, we fail to reject it. Below will be the attached code modelling the above methodology. There will also be an accompanying graph for visual representation.

### Code:

```python
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(1)

# Number of simulations for the hypothesis test
number_of_simulations = 10000

# Sample size
sample_size = 80

# Observed number of correct guesses
observed_correct_guesses = 49

# Observed test statistic: sample mean
observed_test_statistic = observed_correct_guesses / sample_size

# Initialize an array to store simulated sample means
simulated_means = np.zeros(number_of_simulations)

# Generate synthetic samples under the null hypothesis (H0: mu = 0.5)
for i in range(number_of_simulations):
    # Generate random guesses assuming the null hypothesis is true
    random_guesses = np.random.choice([0, 1], size=sample_size, replace=True)
    simulated_means[i] = random_guesses.mean()

# Calculate how many simulated means are as or more extreme than the observed test statistic
extreme_test_stats = np.abs(simulated_means - 0.5) >= np.abs(observed_test_statistic - 0.5)

# Calculate the p-value
p_value = extreme_test_stats.sum() / number_of_simulations

# Output the results
print("Number of Simulations:", number_of_simulations, "\n",
      "Number of simulated test statistics (under H0)\n",
      "that are as or more extreme than the observed test statistic:", extreme_test_stats.sum(), "\n",
      "Two-tailed p-value\n(= simulations as or more extreme than observed / total simulations):", p_value)

print("\n\n")

# Display a DataFrame showing which simulated test statistics are "as or more extreme"
results_df = pd.DataFrame({
    '(Simulated) Test Statistic': simulated_means,
    '>= |' + str(observed_test_statistic) + '|?': ['>= |' + str(observed_test_statistic) + '|?'] * number_of_simulations,
    '"as or more extreme"?': extreme_test_stats 
})

print(results_df.head(10))  # Display the first few rows of the DataFrame
```
