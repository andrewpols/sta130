## Analysis #1
**Research Question:** Is there an association between the proportion of friends who share the same ethnicity as a person and their reported body image?  

**Variables:**
- **Predictor:** *CONNECTION_close_friends_same_ethnicity* — an ordinal variable assessing the proportion of friends who share a respondent's ethnicity.
- **Outcome:** *PSYCH_body_self_image_questionnaire_think_unattractive* — an ordinal variable measuring respondents’ self-perception of body image.

**Rationale for Variable Selection:** This analysis explores whether social alignment within friendship groups impacts self-image. Research indicates that social similarities can influence individual self-esteem and identity, making this a significant area to examine.

**Visualization:** A stacked bar plot will visualize patterns in body image across different proportions of ethnically similar friends. This format, especially when based on percentages, allows for clear comparisons of multiple categories, making it easier to observe trends and differences in body image perceptions.

**Summary Statistics:** Descriptive statistics, such as means and frequencies, will accompany the visualization to provide a clearer understanding of the distribution and central tendency of body image perceptions across the categories.

**Analysis:** A Chi-squared test will evaluate the association between the proportion of friends sharing the same ethnicity and body image. An alpha significance level of 0.01 will strengthen our threshold for evidence against the null hypothesis, since this test does not provide exact measures. Note that Fisher-exact tests are not ideal for larger contingency tables with many categories.

**Assumptions:** Independence of observations, expected frequencies of at least 5, a sufficiently large sample size, and categorical data.

**Hypotheses and Possible Results:**
- **$H_0$** There is no association between friend group ethnicity proportion and body image.  
- **$H_A$:** There is an association.

**Predicted Results:**
- There is an association between the proportion of friends who share the same ethnicity and reported body image, suggesting that social belonging may influence perceptions.

**Relevance:** A significant association ($p < 0.01$) would indicate that ethnic similarity within social groups impacts self-perception. A non-significant result  $(p ≥ 0.01)$ would imply that ethnic similarity does not influence body image.

**Ethical Considerations:** This analysis will handle the sensitive nature of self-reported body image and ethnic identity with care, ensuring confidentiality and responsible interpretation of observed associations.

## Analysis #2

**Research Question:** Is there a difference in work hours per week between individuals who are parents and those who are not?

**Variables:**
- **Predictor:** *GEO_housing_live_with_children* — a numeric variable indicating the number of children an individual lives with, which we will use to create a binary indicator variable.
- **Outcome:** *WORK_hours_per_week* — a numeric variable representing weekly work hours.

**Rationale for Variable Selection:** Analyzing differences in work hours based on parental status provides insights into work-life balance and the pressures family responsibilities impose on time allocation.

**Visualization:** A box plot will visualize the distribution of work hours among parents and non-parents, highlighting central tendencies and variances. Histograms and Kernel Density Estimates (KDEs) will help identify differences in modality between groups.

**Analysis:** We will implement a double bootstrap approach to compare the means of work hours between parents (those living with children) and non-parents (those not living with children). This involves bootstrapping both groups in parallel and calculating the difference in means for each bootstrap sample. The results will include a 95% confidence interval to assess the statistical significance of the differences observed.

**Assumptions:** Independence of observations and the validity of the bootstrap method, which does not assume normality of the data (as well as unbiased sampling).

**Hypotheses and Possible Results:**

Based on our 95% confidence interval, we can either reject the null, or fail to reject it based on the following hypotheses:
- **$H_0$:** $\mu_\textit{parents} = \mu_\textit{non-parents}$. There is no difference in work hours per week between parents and non-parents. 
- **$H_A$:**  $\mu_\textit{parents} \neq \mu_\textit{non-parents}$. There is a difference in work hours per week between parents and non-parents.

**Predicted Results:**
- Parents work more hours per week than non-parents due to the financial demands of raising children.

**Relevance:** A significant positive difference ($H_A$), which could be inferred based on the confidence interval, would suggest that parents work more than non-parents on average, informing support policies for work-life balance, such as flexible schedules and childcare support. Conversely, a non-significant result ($H_0$) would indicate that other factors influence time allocation, guiding future research on workplace dynamics.

**Ethical Considerations:** This analysis acknowledges the complexity of family dynamics. Results will be interpreted sensitively, recognizing that work hours may reflect economic necessity, personal choice, or systemic pressures, rather than solely parental responsibilities.

## Analysis #3 
**Research Question:** Is there an association between the frequency of church attendance and self-rated mental health?  

**Variables:**
- **Predictor:** *CONNECTION_activities_church_last* — a nominal variable describing when someone last attended church, temple, mosque, etc.
- **Outcome:** *WELLNESS_self_rated_mental_health* — an ordinal variable assessing self-rated mental health.

**Rationale for Variable Selection:** Investigating the relationship between religious participation and mental health aligns with theories suggesting that community engagement may support mental well-being.

**Visualization:** A stacked bar plot will show the distribution of mental health ratings across church attendance categories, facilitating clear visual comparisons. Specifically percentage-based stacked bar plots will highlight differences in mental health ratings associated with varying levels of church attendance.

**Summary Statistics:** Descriptive statistics will accompany the visualization to summarize mental health ratings within each attendance category, providing insights into the central tendencies and variability of responses.

**Analysis:** A Chi-squared test will determine if there is a statistically significant association between church attendance and self-rated mental health. Since this analysis involves an estimate, an alpha level of 0.01 will ensure significant findings will provide strong evidence against the null hypothesis.

**Assumptions:** Independence of observations, expected frequencies of at least 5, sufficient sample size, and categorical data.

**Hypotheses and Possible Results:**
- **$H_0$:** There is no association between church attendance and mental health.  
- **$H_A$:** There is an association.

**Predicted results:**
- There is an association between frequency of church attendance and self-rated mental health, possibly linked to social support from religious communities.

**Relevance:** A significant association ($p < 0.01$) would suggest that church attendance correlates with well-being. A non-significant result ($p \geq 0.01$) would imply that church attendance does not significantly affect mental health.

**Ethical Considerations:** This analysis respects the diversity of spiritual beliefs, ensuring findings do not imply universal benefits of religious practices. We aim to explore potential associations without advocating for increased religious participation.
