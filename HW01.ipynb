# 5. Attribute vs. Method

Chat-GPT Log: https://chatgpt.com/share/f88357c4-9dc7-4a07-b652-af3f2fa64787

## Attributes:

An attribute, like <code>td.shape</code>, is a **property of an object, holding some form of data** that can be accessed without performing any action on an object. An attribute does not use parantheses because it doesn't perform anything **on** the object, it simply gives details **about** the object. For example, <code>td.shape</code> provides the structure of the pandas DataFrame object by returning the number of rows and columns, without performing anything on the DataFrame.


## Method:

A method, like <code>td.describe()</code>, is a **function that performs an operation on an associated object, and may return a result of the object's data**. Since a method is a type of function (more specifically, a function associated with an object), they require parentheses when called (regardless of if they take arguments or not).

### Difference Between a Method and a Function
The main takeaway between a method and a function is that all methods are functions, but not all functions are methods; that is, a method is a specific type of function.

Functions are definitions of reusable code that aren't tied to any object, while a method has an associated object that it performs operations on.

### Difference Between a Function in Mathematics and Programming
Mathematical functions are well-defined in the sense that an input will **always** lead to the same output; semantically, we would call such a function **deterministic**. 

Programming functions **do not** guarantee that an input will **always** lead to the same output. One instance of this could be caused by "side effects," which could occur when passing a global variable into a function and then modifying that variable. There is also an element of uncertainty in programming functions, with programming languages being able to produce random integers that may be used to manipulate an input, leading to a random output (although, we **technically** cannot produce true randomness in computers).

  - **Math Functions**:Mathematical Functions: These are relations that assign each input exactly one output. Examples include f(x) = x^2 or g(x) = sin(x). They describe relationships and operations within mathematical contexts.

# 6. <code>td.describe()</code> Summary Statistics Definitions

Chat-GPT Log: https://chatgpt.com/share/66e3b26e-0320-8008-b75c-c149f1715eb9

1. Count: The total number of non-null observations for each variable. In Statistics, this would represent the sample size of a dataset.

<br>

2. Mean: Known as the artithmetic mean, this statistic is the sum of all observations divided by the count ($n$). Given a variable $x$, its arithmetic mean ($\bar{x}$) is denoted by:

$$\large\bar{x} = \dfrac{1}{n}\displaystyle\sum_{i=1}^{n} x_i$$

<br>

3. std: Known as Standard Deviation ($s$), this statistic measures the "spread" of the data around the mean. It is the square root of the average of the squared deviations from the mean (the square root of the square ensures only positive results):

$$\large s = \sqrt{\dfrac{1}{n-1}\displaystyle\sum_{i=1}^{n} (x_i - \bar{x})^2}$$

<br>

4. Min: The minimum (smallest) value in a dataset:
$$\large Min = min(X_1, X_2, \cdots, X_n)$$

<br>

5. 25%: Known as the first quartile ($Q_1$), this statistic measures the value below which a quarter of the data points are found when arranged in increasing order. It measures the lower portion of the data's spread.

<br>

6. 50%: This represents the median ($Q_2$), or the value in the "middle" of the sorted dataset. So, half of the datapoints lie below this value, while the other half lie above this value.

<br>

7. 75%: The third quartile ($Q_3$) marks the value at which 75% of the datapoints lie below, meaning this value is the upper bound of three quarters if the data.

<br>

8. Max: The maximum (greatest) value observed in the dataset; the upper boundary of the set.

$$\large Max = max(X_1, X_2, \cdots, X_n)$$

## GPT Session Summary

Here’s a summary focusing on the definitions:

1. **Summary Statistics**:
   - **`count`**: Number of non-null observations.
   - **`mean`**: Average value of the data.
   - **`std`**: Standard deviation, measuring the dispersion of the data.
   - **`min`**: Minimum value in the data.
   - **`25%`**: 25th percentile (first quartile), the value below which 25% of the data fall.
   - **`50%`**: Median (50th percentile), the middle value of the data.
   - **`75%`**: 75th percentile (third quartile), the value below which 75% of the data fall.
   - **`max`**: Maximum value in the data.

# 7. Missing Data: <code>td.dropna()</code> and <code>del td['col']</code>

Chat-GPT Log: https://chatgpt.com/share/66e3b26e-0320-8008-b75c-c149f1715eb9

<code>td.dropna()</code> deletes an entire row of data if a single data point within that row is missing a variable; that is, if any of the associated columns fo a row are missing. By default, the axis of this method is 0, but changing that axis to 1 will cause the method to delete columns of data given an associated row that is null: 
<code>td.dropna(axis=1)</code>.

On the other hand, <code>del td['col']</code> deletes an entire column of data upon request.
 
1. Using <code>td.dropna()</code> would be preffered over <code>del td['col']</code> in a case where there is more missing data in a given row rather than a given column. This preserves as much valuable data as possible by only deleting the data which has missing values associated with it. For example, if a row of data was missing a passenger's name, it wouldn't make sense to delete the entire column of passenger names because of one missing data point. Instead, we would use <code>td.dropna()</code> to drop the row with the missing name.

<br>

2. Contrarily to 7.1, <code>del td['col']</code> would be useful if there were more data missing in a given column than a row. This way, the unuseful data would be deleted along a column without affecting the rest of the data in a row. For example, in the titanic dataset, there are 687/891 cabin IDs missing. It wouldn't make sense to delete 687 rows of data to account for this. Instead, it would be wiser to delete the column of cabin IDs entirely and prevent any further valuable data from being deleted across a row. <br><br> This is further warranted if the column in question is not useful. The cabin ID column yields no statistical significance and any associated rows should not be deleted because of it.

<br>

3. In order to prevent as much unecessary data loss as possible, using <code>del td['col']</code> first would help <code>td.dropna()</code> target useful null values. In a similar premise to 7.2, it would make sense to delete the cabin IDs first. That way, there wouldn't be as many rows deleted from null values. For example, if there was just two rows, each with a missing value somewhere, deleting the cabin column first would not only prevent over 680 rows from being deleted, but would then allow for <code>td.dropna()</code> to target the rest of the null rows. Also, as stated in 7.2, insignificant columns like the cabin ID can be deleted without consequence because there isn't much stastical analysis you can perform on such data, and would allow for focusing <code>td.dropna()</code> on relevant data.

<br>

4. As explored earlier, it would make sense to first delete the Cabin column, as there is no statistical significance to this column, and just 204 non-null values would not be a big enough sample size to work with anyway. This prevents <code>td.dropna()</code> from deleting 687 rows of useful data; this would delete 177 rows of missing ages and 2 rows of embarked status.<br>**Note**: Deleting the ages column would not be smart as age is a useful variable for statistical analysis, and there isn't a relatively large amount of data missing.

````python
dropped_td = td.copy() # to compare before and after

if ('Cabin' in td.columns):
    del dropped_td['Cabin']

dropped_td = dropped_td.dropna()

dropped_td.head(20)
````

````python
dropped_td.describe()
````

````python
print(dropped_td.isna().sum()) # Ensure no null values
````

## GPT Session Summary

1. **`dropna()` Function**:
   - The `dropna()` function removes missing values from a DataFrame or Series. It can drop rows or columns with NaNs, and its parameters include `axis` (rows/columns), `how` (criteria for dropping), `thresh` (minimum number of non-NaN values), `subset` (specific columns to consider), and `inplace` (whether to modify the original DataFrame).

2. **`del df['col']` vs. `dropna()`**:
   - `del df['col']` removes a specific column from the DataFrame, while `dropna()` removes rows or columns with NaNs. Using `dropna()` might be preferred if you want to retain as much data as possible by removing only incomplete rows or columns. 

3. **Use Case for `del df['col']` Before and With `dropna()`**:
   - Deleting irrelevant columns before using `dropna()` helps prevent unnecessary removal of rows. For example, removing columns with high proportions of missing data first ensures `dropna()` only affects relevant columns, thereby preserving valuable data and improving data processing efficiency.
